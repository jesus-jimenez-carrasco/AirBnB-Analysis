```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
import matplotlib as plt
```

```python
conditions_df = pd.read_parquet("airbnb_madrid_conditions.parquet")
host_df = pd.read_parquet("airbnb_madrid_host.parquet")
location_df = pd.read_parquet("airbnb_madrid_location.parquet")
property_df= pd.read_parquet("airbnb_madrid_property.parquet")
reviews_df = pd.read_parquet("airbnb_madrid_reviews.parquet")
```

```python
df = pd.merge(left=conditions_df, right=host_df, on="id", how="outer")
df_ = pd.merge(left=df, right=location_df, on="id", how="outer")
df_2 = pd.merge(left=df_,right=property_df,on="id",how="outer")
df_3 = pd.merge(left=df_2, right=reviews_df,on="id",how="outer")
df_final = df_3
df_final = df_final[df_final["price"]<= 350]
```

```python
# Creo Dataframes añadiendo el ID y el precio
df_host = pd.read_parquet("airbnb_madrid_host.parquet")
df_conditions = pd.read_parquet("airbnb_madrid_conditions.parquet")
df_location = pd.read_parquet("airbnb_madrid_location.parquet")
df_property = pd.read_parquet("airbnb_madrid_property.parquet")
df_reviews = pd.read_parquet("airbnb_madrid_reviews.parquet")

df_host_price = df_final[["id", "price"]].merge(df_host, on="id", how="left")
df_conditions_price = df_final[["id", "price"]].merge(df_conditions, on="id", how="left")
df_location_price = df_final[["id", "price"]].merge(df_location, on="id", how="left")
df_property_price = df_final[["id", "price"]].merge(df_property, on="id", how="left")
df_reviews_price = df_final[["id", "price"]].merge(df_reviews, on="id", how="left")
```

# Hipótesis 1 - La localización impacta directamente en el precio. A más cerca del centro, mayor precio

```python
# Distancia media Aeropuerto
dist_media_airport = df_final.groupby("neighbourhood_group")["dist_km_airport"].median().sort_values()
print(dist_media_airport)
dist_media_total_ = dist_media_airport.median()
print(f"De media, los inmuebles se encuentran a {dist_media_total_} kilómetros del aeropuerto")
is_top10_ = ["Barajas", "San Blas - Canillejas", "Hortaleza", "Ciudad Lineal", "Vicálvaro", "Chamartín", "Salamanca", "Moratalaz", "Tetuán", "Retiro", "Chamberí"]
df_final["is_top10_"] = df_final["neighbourhood_group"].apply(lambda x: x in is_top10_)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
ax = plt.gca()
sns.boxplot(data=df_final,
            x="is_top10_",
            y="price")
ax.set_ylim(0, 350)
ax.set_yticks([0,50,100,150,200,250,300,350])
plt.show()
```

# Hipótesis 3 - Las Reviews impactan en el precio del inmueble

```python
#Existe un 10% de valores nulos en el dataframe, los cuales no quiero eliminar pero les asigno el valor de 0
df_reviews_price = df_reviews_price.fillna(value=0)

print(df_reviews_price.isnull().sum())
df_reviews_price.head()


```

```python
df_reviews_price["number_of_reviews"] = df_reviews_price["number_of_reviews"][df_reviews_price["number_of_reviews"] > 0]
```

```python
median_reviews = df_reviews_price["number_of_reviews"].median()
median_reviews_mnth = df_reviews_price["reviews_per_month"].median()
median_reviews_ltm = df_reviews_price["number_of_reviews_ltm"].median()
total_scores = df_reviews_price["review_scores_cleanliness"] + df_reviews_price["review_scores_checkin"] + df_reviews_price["review_scores_communication"] + df_reviews_price["review_scores_location"] + df_reviews_price["review_scores_value"]
df_reviews_price["total_scores"] = total_scores
print(f"De media los inmuebles reciben un total de {median_reviews} reviews")
print(f"De media los inmuebles reciben mensualmente una media de {median_reviews_mnth} reviews")
print(f"Los inmuebles han recibido {median_reviews_ltm} reviews en los últimos 12 meses")
```

```python
#Number of Reviews Vs Price

plt.figure(figsize=(10, 7))
ax = plt.gca()
sns.scatterplot(data=df_reviews_price,
            x="number_of_reviews",
            y="price",
            alpha=0.3)
ax.set_ylim(0,350)
ax.set_yticks([0,50,100,150,200,250,300,350])
plt.show()

spearman_corr_reviews = df_final[["number_of_reviews", "price"]].corr(method="spearman")
print(spearman_corr_reviews)
```

```python

df_reviews_price["total_scores"] = df_reviews_price["review_scores_cleanliness"] + df_reviews_price["review_scores_checkin"] + df_reviews_price["review_scores_communication"] + df_reviews_price["review_scores_location"] + df_reviews_price["review_scores_value"]

# Total Scores en Reviews Vs Precio
plt.figure(figsize=(10, 7))
ax = plt.gca()
sns.scatterplot(data=df_reviews_price,
            x="total_scores",
            y="price",
            alpha=0.3,
            hue="price")
ax.set_ylim(0, 350)
ax.set_yticks([0, 50, 100, 150, 200, 250, 300, 350])
plt.show()
corr_totalscores_price = df_reviews_price[["total_scores", "price"]].corr(method="spearman")
print(corr_totalscores_price)
```

# Hipótesis 4 - El tipo de Propiedad Impacta en el Precio

```python
# Descarto utilizar las columnas de Property_type y Room_type ya que solo existe una opcion: son apartamentos que se alquilan enteros
# Creo una nueva columna con el numero total de habitaciones (Baños, Dormitorios, etc)
df_property_price["total_rooms"] = df_property_price["bathrooms"] + df_property_price["bedrooms"]
df_property_price.head()

# Sumo las Amenities (Dummies) y me baso en el siguiente baremo:
# Low Equipment <= 2,5 ptos. Mid_Equipment == 2,5 ptos. High Equipment > 2,5 ptos

total_amenities = df_property_price["amenities_wifi_internet"] + df_property_price["amenities_tv"] + df_property_price["amenities_air_conditioning"] + df_property_price["amenities_heating"] + df_property_price["amenities_kitchen"]
df_property_price["total_amenities"] = total_amenities

df_property_price["low_equip"] = (df_property_price["total_amenities"] < 2.5).astype(int)
df_property_price["mid_equip"] = (df_property_price["total_amenities"] == 2.5).astype(int)
df_property_price["high_equip"] = (df_property_price["total_amenities"] > 2.5).astype(int)

conditions = [
    df_property_price["total_amenities"] < 3,
    df_property_price["total_amenities"] == 3,
    df_property_price["total_amenities"] > 3
]
choices = ["low_equip", "mid_equip", "high_equip"]

df_property_price["equip_level"] = np.select(conditions, choices, default="unknown")

plt.figure(figsize=(10, 7))
ax = plt.gca()
sns.boxplot(data=df_property_price,
            x="equip_level",
            y="price")
ax.set_ylim(0,350)
ax.set_yticks([0,50,100,150,200,250,300,350])
plt.show()

corr_amenities = df_property_price[["total_amenities", "price"]].corr(method="spearman")
print(corr_amenities)




```
